```python
%sql

-- metadata database and tables to track processed batches and their status. This is used for monitoring, debugging, and ensuring idempotency in the ETL pipeline.
CREATE DATABASE IF NOT EXISTS metadata;

CREATE TABLE metadata.processed_batches (
    pipeline_name STRING NOT NULL,
    batch_id STRING NOT NULL,
    source_path STRING,
    status STRING,
    processed_at TIMESTAMP,
    CONSTRAINT pk_batches PRIMARY KEY (pipeline_name, batch_id)
)
USING DELTA;
```
